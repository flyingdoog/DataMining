```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('./data/noise.txt', sep="\t")
data
```

```python
# For the objects without labels, we just remove the data objects.
data_remove_unlabel = data[data['type']!='?']
data_remove_unlabel = data_remove_unlabel.astype({'type': 'int64'})
```

```python
# Fill with constant, such as 0
fill_0 = data_remove_unlabel.copy()
fill_0.replace({'?': 0})
```

```python
# fill in with attribute mean
def fill_with_mean(df,col):
    # get the records with values
    df_with_values = df[df[col]!='?']
    df_with_values = df_with_values.astype({col: 'float'})
    df_mean = df_with_values.mean(axis = 0)
    df_return = df.copy()
    df_return[col].replace({'?': int(df_mean[col])},inplace=True)
    return df_return
fill_with_mean(data_remove_unlabel,'size')
```

```python
# fill in with attribute mean with records with the same label
def fill_with_label_mean(df,col,label_col='type'):
    # get the records with values
    df_with_values = df[df[col]!='?']
    labels = df_with_values[label_col].unique()
    df_returns = []
    for label in labels:
        df_label = df[df[label_col]==label]
        df_label = fill_with_mean(df_label,col)
        df_returns.append(df_label)
    df_return = pd.concat(df_returns).sort_index()
    return df_return
fill_with_label_mean(data_remove_unlabel,'size')
```
